```python
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get("https://www.bajajfinservhealth.in/search?source=In_Clinic_Doctors&inclinic=true&order_by=distance&page=1")
```

```python
soup = BeautifulSoup(req.content, "html.parser")
```

```python
docdiv = soup.find('div', class_='css-gcv1rn').find_all('div', class_='css-1koyru1')
```

```python
for i in range(1,10):
    req = requests.get("https://www.bajajfinservhealth.in/search?source=In_Clinic_Doctors&inclinic=true&order_by=distance&page="+str(i))
    soup = BeautifulSoup(req.content, "html.parser")
    docdiv = soup.find('div', class_='css-gcv1rn').find_all('div', class_='css-1koyru1')
    for doclist in docdiv:
        print(doclist.find('h2', class_='css-hj0zi4').text)
        print(doclist.find('p').text)
        print(doclist.find_all('span')[1].text)
        print(doclist.find('p', class_ = 'css-156w1ve').text)
```
